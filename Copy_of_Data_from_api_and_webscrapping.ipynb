```python
students=[]
# print(soup.find_all('article',class_='student'))    #  find_all finds all the matching object while find gets only first matcching object
# print(type(soup.find_all('article',class_='student')))
for student in soup.find_all('article',class_='student'):
  name=(student.find_all('p',class_='student-name')[0].text.replace('Name:','').strip()) # no need of find_all when we have specified class name
  roll=(student.find('p',class_='student-roll').text.replace('Roll No:','').strip())
  age=(student.find('p',class_='student-age').text.replace('Age:','').strip())
  email=(student.find('p',class_='student-email').text.replace('Email:','').strip())
  # print(student.h3.text)
  # print((student))
  # print(student.p.text)
  students.append({'name':name,'roll':roll,'age':age,'email':email})
students
```
