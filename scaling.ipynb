```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# STANDARDIZATION

```python
data=pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
data.sample()
```

```python
data=data.iloc[:,2:]
```

```python
data.sample()
```

```python
x=data.iloc[:,0:2]
y=data.iloc[:,-1]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train
```

```python
# sc.fit finds mean and standard daviation for all the columns of x_train
# and on the basis of that it transforms the data both one train and test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train_scaled=sc.transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
sc.mean_
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_test_scaled
```

```python
x_train_scaled.describe()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))
ax1.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],color='blue',label='AFTER SCALING')
ax2.scatter(x_train['Age'],x_train['EstimatedSalary'],color='green',label='BEFORE SCALING')
ax1.legend()
ax2.legend()
# only ranges of data arre changed
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'], ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_train_scaled['Age'], ax=ax2)
sns.kdeplot(x_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

#Comparing distribution

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(x_train['Age'], ax=ax1)

# after scaling
ax2.set_title('Age Distribution After Standard Scaling')
sns.kdeplot(x_train_scaled['Age'], ax=ax2)
plt.show()
```

```python
df=pd.read_csv("/content/wine_data.csv",names=['Class Label','Alcohol','Malic Acid','z','z1','z3','z2','z4','z5','z6','z7','z8','z9','z10'])
```

```python
df.head()
```

```python
df=df.iloc[:,:3]
df.columns=['Class Label','Alcohol','Melic Acid']
```

```python
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.scatterplot(x=df['Alcohol'],y=df['Melic Acid'],hue=df['Class Label'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(['Class Label'],axis=1),df['Class Label'],test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(x_train)
```

```python
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_train_scaled.describe()
```

```python
np.round(x_train.describe())
```

```python
y_train
```

```python
fig ,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(x_train_scaled['Alcohol'],x_train_scaled['Melic Acid'],c=y_train,label='After scaling')
ax1.set_title('After scaling')
ax2.scatter(x_train['Alcohol'],x_train['Melic Acid'],c=y_train,label='Before scaling')
ax2.set_title('Before scaling')
plt.show()
```

```python
fig ,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title('After scaling')
sns.kdeplot(x_train_scaled['Alcohol'],ax=ax1)
sns.kdeplot(x_train_scaled['Melic Acid'],ax=ax1)
ax2.set_title('Before scaling')
sns.kdeplot(x_train['Alcohol'],ax=ax2)
sns.kdeplot(x_train['Melic Acid'],ax=ax2)
plt.show()
# distribution shape may change but mist cases will not change
```
