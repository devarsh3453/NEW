```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Titanic_Imputer.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head(11)
```

```python
df.isna().sum()
```

```python
X=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
x=[]
for val in ['distance','uniform']:
  for i in range (1,10):
    knn = KNNImputer(n_neighbors=i,weights=val)

    X_train_trf = knn.fit_transform(X_train)
    X_test_trf = knn.transform(X_test)
    lr = LogisticRegression()

    lr.fit(X_train_trf,y_train)

    y_pred = lr.predict(X_test_trf)
    x.append(accuracy_score(y_test,y_pred))
    print(f" K value {i:3}    Weight  {val:<4}  Accuracy is  {accuracy_score(y_test,y_pred):^6}")
```

```python
arr=np.array(x)
arr.max()
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf,y_train)

y_pred = lr.predict(X_test_trf)

accuracy_score(y_test,y_pred)
```
