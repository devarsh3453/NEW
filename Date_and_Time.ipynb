```python
date=pd.read_csv('orders.csv')
time=pd.read_csv('messages.csv')
```

```python
#converting datatype
date['date']=pd.to_datetime(date['date'])
time['date']=pd.to_datetime(time['date'])
```

```python
date.info()
```

```python
date['Date_Y']=date['date'].dt.year  #year
date['Date_M']=date['date'].dt.month  # Month
date['Date_M_name']=date['date'].dt.month_name() # Month name
date['Date_D']=date['date'].dt.day # day
date['Date_DofW']=date['date'].dt.dayofweek # number of day in a week
date['Date_D_name']=date['date'].dt.day_name() # day name
date['Date_W']=date['date'].dt.isocalendar().week # week number
date['Date_Q']=date['date'].dt.quarter # quater
date['Date_IS_WEEKEND']=np.where(date['Date_D_name'].isin(['Saturday','Sunday']),1,0)
date['Date_Sem']=np.where(date['Date_Q'].isin([1,2]),1,2)
date.sample(10)
```

```python
#Time Elapsed info
import datetime as dt
today=dt.datetime.today()
today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
#  Months passed
np.round((today -date['date']) / np.timedelta64(1,'m'),0)
```
