### opening a csv from an url (from server)

```python
import requests
from io import StringIO

url="https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers={"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req=requests.get(url,headers=headers)
data=StringIO(req.text)

df=pd.read_csv(data)
df
```

### passing a list to set as column names

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',names=['i','j','k','l','m','n','o','p','q','r'])
# if give extra column name than create a extra column with NAN values
df
```

###Index_col making a column as index

```python
pd.read_csv('/content/sample_data/california_housing_test.csv',index_col='longitude')
```

### header parameter

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',header=1)
# makes 1 st row as header if header is in first row
df
```

### use_cols parameters

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',usecols=['longitude','latitude','housing_median_age'])
# only mentioned columns are imported
df
```

### Squeeparameter

### Skiprow and nrows

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',skiprows=[1,2,3,4,5])
# skip specific rows to skip and we can write our lambda function here
df
```

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',skiprows=(lambda x: x in [0,40] ))
df
```

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',nrows=5)
# only get specified number of rows  (used for loading small amount of data form huge set)
df
```

### skipping the bad lines

```python
df = pd.read_csv("/content/sample_data/california_housing_test.csv", on_bad_lines="skip")   # skip malformed lines
df = pd.read_csv("/content/sample_data/california_housing_test.csv", on_bad_lines="warn")   # warn then skip
df = pd.read_csv("/content/sample_data/california_housing_test.csv", on_bad_lines="error")  # error on that line
# skip the lines which have more values that mentioned columns
```

### using dtype

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',dtype={'housing_median_age':int})
df
```

### date-times

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',parse_dates=['longitude'])
# panda treat date as string while reading so we convert it back to date-time formate
df
```

### Applying some transformations on the column names

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',converters={'longitude':lambda x:x.upper()})
df
```

### na_values parameters

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv',na_values=['@','!']) # consider mentioned values as nan
```

### loading huge dataset in chunks

```python
dff=pd.read_csv('/content/sample_data/california_housing_test.csv',chunksize=500)
for i in dff:
  print(i)
```
