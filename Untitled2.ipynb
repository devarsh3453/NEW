```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/placement.csv')
# using logistic regrration it makes a line which will saperate the two classes LR is good for here. line will be desided by algo.
```

```python
df
```

```python
df.head()
df.info()
```

```python
df=df.iloc[:,1:]  # preprocessing the data remove first column not nessecery so remove it
df.shape
```

```python
#steps

#0. Preprocess + EDA + Feature Selection  (missing values , removing columns , outliers)
#1. Extract input and output columns
#2. Scale the values
#3. Train test split (Cross validation ) (we hide some data and call it testing set for testing which is not provided for trainign)
#4. Train the model
#5. Evaluate the model/ model selection
#6. Deploy the model
```

```python
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split  # train,test,split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
from sklearn.preprocessing import StandardScaler   # scaling
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train)   # traing the model
```

```python
y_pridict=clf.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score  # for accuracy
accuracy_score(y_test,y_pridict)
```

```python
# plotting the desicion boundary
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train.values,clf=clf,legend=2)
```
