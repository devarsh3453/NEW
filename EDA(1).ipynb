#Categirical data

Countplots ?? asks how many of each category


```python
sns.countplot(x='Survived',data=df) # plots freq of each category
df['Survived'].value_counts()
# df['Survived'].value_counts().plot(kind='bar') # does the same thing
```

```python
sns.countplot(x='Pclass',data=df)
df['Pclass'].value_counts().plot(kind='pie')
```

```python
sns.countplot(x='Sex',data=df)
df['Sex'].value_counts()
```

PI chart shows %

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```
