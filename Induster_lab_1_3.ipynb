```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simulate sensor data
np.random.seed(19)  # For reproducibility

def simulate_sensor_data(sensor_type):
    n_points = 1100
    if sensor_type == 'temperature':
        data = 20 + 5 * np.random.randn(n_points)  # Mean 30, std 5
    elif sensor_type == 'pressure':
        data = 1100 + 20 * np.random.randn(n_points)  # Mean 1100, std 20
    else:
        raise ValueError("Unknown sensor type")
    return data
```

```python
temperature_data = simulate_sensor_data('temperature')
pressure_data = simulate_sensor_data('pressure')
```

```python
temperature_data
```

```python
temperature_avg = np.mean(temperature_data)
pressure_avg = np.mean(pressure_data)
temperature_peak = np.max(temperature_data)
pressure_peak = np.max(pressure_data)
temperature_min = np.min(temperature_data)
pressure_min = np.min(pressure_data)
```

```python
print(f"Temperature - Avg: {temperature_avg:.2f}, Peak: {temperature_peak:.2f} and Min : {temperature_min:.2f}")
print(f"Pressure - Avg: {pressure_avg:.2f}, Peak: {pressure_peak:.2f} and Min : {pressure_min:.2f}")
```

```python
# Create a DataFrame
data = {
    'Temperature': temperature_data,
    'Pressure': pressure_data
}
df = pd.DataFrame(data)

# Specify the path where you want to save the CSV file
csv_file_path = os.path.join(os.getcwd(), 'sensor_data_2.csv')

# Save the DataFrame to a CSV file at the specified path
df.to_csv(csv_file_path, index=False)

# Confirm that the file has been saved
print(f"CSV file saved at: {csv_file_path}")
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(temperature_data, label='Temperature')
plt.axhline(y=temperature_avg, color='r', linestyle='--', label='Average')
plt.title('Temperature Sensor Data')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(pressure_data, label='Pressure')
plt.axhline(y=pressure_avg, color='r', linestyle='--', label='Average')
plt.title('Pressure Sensor Data')
plt.xlabel('Time')
plt.ylabel('Pressure (hPa)')
plt.legend()
```
