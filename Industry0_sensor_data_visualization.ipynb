```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta


# Simulation parameters
num_samples = 100
start_time = datetime.now()

# Generate timestamps
timestamps = [start_time + timedelta(seconds=i*10) for i in range(num_samples)]

# Simulate temperature data (in Celsius)
temperature = np.random.uniform(20, 30, num_samples)

# Simulate pressure data (in hPa)
pressure = np.random.uniform(1000, 1020, num_samples)

# Create a DataFrame
data = pd.DataFrame({
    'Timestamp': timestamps,
    'Temperature (°C)': temperature,
    'Pressure (hPa)': pressure
})

# Save to CSV file
data.to_csv('sensor_data.csv', index=False)

print("Sensor data simulated and saved to 'sensor_data.csv'")
print(data.head())
```

```python
import matplotlib.pyplot as plt
# Convert Timestamp to datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
# Create subplots for temperature and pressure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot temperature data /// line plot how temperature and pressure change over time.
ax1.plot(data['Timestamp'], data['Temperature (°C)'], color='red', linewidth=2)
ax1.set_xlabel('Time')
ax1.set_ylabel('Temperature (°C)')
ax1.set_title('Temperature Over Time')
ax1.grid(True)

# Plot pressure data // line plot
ax2.plot(data['Timestamp'], data['Pressure (hPa)'], color='blue', linewidth=2)
ax2.set_xlabel('Time')
ax2.set_ylabel('Pressure (hPa)')
ax2.set_title('Pressure Over Time')
ax2.grid(True)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
